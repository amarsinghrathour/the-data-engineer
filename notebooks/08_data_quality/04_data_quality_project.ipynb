# Data Quality Project: Reliable Pipeline Layer


## Project Goal

Build a data pipeline with a dedicated validation layer.


## Features

- Validate incoming data
- Reject bad records
- Log issues
- Track metrics


## Suggested Build

1. Extract data from your source.
2. Validate all rows before load.
3. Store valid rows, reject invalid rows.
4. Write quality logs and summary metrics.


## Interview Line

"I implemented data quality checks in my pipeline to ensure only valid data is processed. Invalid records are logged and excluded, improving reliability."


## What This Adds

👉 Junior: "pipeline works"
👉 Strong engineer: "pipeline is reliable"


## Interview Questions

1. How do you prevent bad data from entering storage?
2. What should happen to rejected records?
3. Which quality checks are mandatory for production?
4. How do you prove quality improvements over time?


## Your tasks

- [ ] Implement validation + rejection path in your pipeline.
- [ ] Log invalid rows to file with reasons.
- [ ] Track and report quality metrics per run.
- [ ] Prepare your interview line with one concrete example.
