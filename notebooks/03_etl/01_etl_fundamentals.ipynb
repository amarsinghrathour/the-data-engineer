# ETL Fundamentals for Data Engineering


## Why ETL Matters

Raw data is messy and useless.

ETL makes data:
- Clean
- Structured
- Usable

Every company:
- Collects raw data
- Transforms it
- Stores it for analytics

👉 This is the core job of a Data Engineer


## What is ETL?

ETL = Extract → Transform → Load

1. **Extract** → Get data (API, files, DB)
2. **Transform** → Clean, filter, enrich
3. **Load** → Store in database or file

## Simple ETL example


```python
import requests
import json


def extract():
    url = "https://jsonplaceholder.typicode.com/posts"
    return requests.get(url).json()


def transform(data):
    return [
        {
            "id": d["id"],
            "title": d["title"].upper()
        }
        for d in data
    ]


def load(data):
    with open("etl_output.json", "w") as f:
        json.dump(data, f, indent=4)


data = extract()
cleaned = transform(data)
load(cleaned)

print("ETL completed")
```

## ETL vs ELT

**ETL:**
- Transform **before** storing

**ELT:**
- Load raw first, transform **in** the warehouse / SQL engine

Modern stacks often **ELT** into cloud warehouses (Snowflake, BigQuery, Redshift) + dbt/SQL transforms.


## Real pipeline thinking

**Example flow:** Orders API → clean → relational store → analytics.

👉 Not just three functions—**data flow**, SLAs, failures, and contracts.
