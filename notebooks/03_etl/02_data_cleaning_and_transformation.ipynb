# Data Cleaning and Transformation in ETL


## Why Data Cleaning Matters

Real-world data is messy:

- Missing fields
- Wrong data types
- Duplicates
- Inconsistent formats

If you don’t clean data:
👉 Your pipeline produces garbage

**Garbage in = garbage out**


## Sample dirty data


```python
data = [
    {"id": 1, "name": "amar", "age": "25"},
    {"id": 2, "name": None, "age": "30"},
    {"id": 3, "name": "NEHA", "age": None},
    {"id": 4, "name": "ravi", "age": "twenty"},
    {"id": 1, "name": "amar", "age": "25"},  # duplicate
]

data
```

## Handling missing values


```python
def clean_missing(data):
    cleaned = []

    for d in data:
        if not d.get("name") or not d.get("age"):
            continue
        cleaned.append(d)

    return cleaned


cleaned = clean_missing(data)
cleaned
```

## Fixing data types

_Using `int(...)`; skip rows that cannot parse (e.g. `"twenty"`) — catch `ValueError` / `TypeError` only._


```python
def fix_types(data):
    result = []

    for d in data:
        try:
            d["age"] = int(d["age"])
            result.append(d)
        except (ValueError, TypeError):
            continue

    return result


fixed = fix_types(cleaned)
fixed
```

## Standardizing data


```python
def standardize(data):
    return [
        {
            "id": d["id"],
            "name": d["name"].title(),
            "age": d["age"],
        }
        for d in data
    ]


standardized = standardize(fixed)
standardized
```

## Removing duplicates


```python
def remove_duplicates(data):
    seen = set()
    unique = []

    for d in data:
        key = (d["id"], d["name"])
        if key not in seen:
            seen.add(key)
            unique.append(d)

    return unique


final_data = remove_duplicates(standardized)
final_data
```

## Full transformation pipeline


```python
def transform_pipeline(data):
    data = clean_missing(data)
    data = fix_types(data)
    data = standardize(data)
    data = remove_duplicates(data)
    return data


transform_pipeline(data)
```

## Real transformation steps

Typical pipeline:

1. Validate data
2. Handle missing values
3. Fix types
4. Normalize formats
5. Remove duplicates
6. Enrich data

👉 **Order matters** — early filters save downstream work.
