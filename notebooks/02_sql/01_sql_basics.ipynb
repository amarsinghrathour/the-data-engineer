## Connect and create table


```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="de_course",
    user="admin",
    password="admin",
)

cur = conn.cursor()

cur.execute(
    """
CREATE TABLE IF NOT EXISTS users (
    id SERIAL PRIMARY KEY,
    name TEXT,
    age INT,
    city TEXT
)
"""
)

conn.commit()
```

## Insert data


```python
cur.execute(
    """
INSERT INTO users (name, age, city)
VALUES
('Amar', 25, 'Bangalore'),
('Ravi', 30, 'Mumbai'),
('Neha', 28, 'Delhi')
"""
)

conn.commit()
```

## Basic queries


```python
cur.execute("SELECT * FROM users")
rows = cur.fetchall()

for row in rows:
    print(row)
```

## Filtering


```python
cur.execute("SELECT * FROM users WHERE age > 26")
print(cur.fetchall())
```

## Aggregations


```python
cur.execute("SELECT COUNT(*), AVG(age) FROM users")
print(cur.fetchall())
```

## Update and delete


```python
cur.execute("UPDATE users SET age = 26 WHERE name = 'Amar'")
cur.execute("DELETE FROM users WHERE name = 'Ravi'")
conn.commit()
```
