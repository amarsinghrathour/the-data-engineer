## Streaming Flow

Producer -> Kafka -> Consumer -> Process -> Output

👉 Data is processed continuously


## Consumer with Processing Logic

```python
from kafka import KafkaConsumer
import json

consumer = KafkaConsumer(
    "test-topic",
    bootstrap_servers='localhost:9092',
    value_deserializer=lambda x: json.loads(x.decode('utf-8')),
    auto_offset_reset='earliest'
)

event_counts = {}

print("Processing stream...")

for message in consumer:
    event = message.value["event"]

    # Count events
    event_counts[event] = event_counts.get(event, 0) + 1

    print("Event:", event)
    print("Counts:", event_counts)
```


## Real-Time Filtering

```python
for message in consumer:
    data = message.value

    if data["event"] == "purchase":
        print("💰 Purchase detected:", data)
```


## Real-Time Aggregation

```python
total_events = 0

for message in consumer:
    total_events += 1
    print("Total events processed:", total_events)
```


## Simulating Alerts

```python
for message in consumer:
    data = message.value

    if data["event"] == "purchase":
        print("🚨 ALERT: Purchase event!")
```


## Real Use Case

User event -> Kafka -> Consumer -> Process -> Store/Alert
