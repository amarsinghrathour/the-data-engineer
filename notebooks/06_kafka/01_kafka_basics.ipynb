# Apache Kafka Basics for Data Engineering


## Why Kafka Matters

Batch pipelines often mean delayed insights and slower reaction time.

Kafka enables:
- Real-time data processing
- Event-driven systems

Common usage examples:
- Uber: ride events
- Netflix: user activity streams
- Payment systems: transaction events


## What is Apache Kafka?

Kafka is a distributed messaging system.

It allows:
- Producers to send data
- Consumers to read data

Data flows continuously in near real time.


## Kafka Concepts

- Topic: stream/category of events
- Producer: writes events to topics
- Consumer: reads events from topics
- Broker: Kafka server that stores and serves data


## Real Flow

User click -> Event -> Kafka -> Processing -> Database

This can happen in milliseconds for many use cases.
