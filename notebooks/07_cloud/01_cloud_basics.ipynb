# Cloud Fundamentals for Data Engineering (AWS)


## Why Cloud Matters

Local systems:
- Limited scale
- Hard to manage

Cloud:
- Scalable
- Reliable
- Pay-as-you-go

👉 All modern data systems run on cloud


## Core Services

Storage -> S3
Compute -> EC2 / Lambda
Database -> RDS
Networking -> VPC

👉 These form your data platform


## Mapping to Data Engineering

Your local pipeline:

API -> Transform -> DB

In cloud (AWS-focused):

API -> EC2 -> S3 -> RDS -> Analytics


## S3 (Simple Storage Service)

- Object storage
- Stores files (CSV, JSON, Parquet)

👉 Used as data lake


## EC2

- Virtual machine
- Run your pipelines

👉 Like your laptop, but in cloud


## RDS

- Managed database (Postgres, MySQL)

👉 Stores structured data
