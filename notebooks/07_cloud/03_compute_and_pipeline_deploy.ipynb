# Deploying Data Pipelines on AWS EC2


## Why Deployment Matters

So far:
- You ran pipelines locally

In real systems:
- Pipelines run on servers
- Accessible 24/7
- Scalable

👉 EC2 lets you run your pipelines in production


## What is EC2?

EC2 = Virtual Machine in AWS

👉 Like your laptop
👉 But always running
👉 Accessible via internet


## Create EC2 Instance

Steps:
1. Go to AWS Console -> EC2
2. Launch instance
3. Choose:
   - Ubuntu
   - t2.micro (free tier)
4. Create key pair
5. Launch


## Connect to EC2

```bash
ssh -i your-key.pem ubuntu@your-ec2-ip
```


## Setup Environment

```bash
sudo apt update
sudo apt install python3-pip -y

pip3 install requests psycopg2-binary boto3 kafka-python
```


## Upload Your Code

Option 1 (recommended):

```bash
git clone https://github.com/your-username/the-data-engineer.git
cd the-data-engineer
```


## Run ETL Pipeline

```bash
python3 projects/batch_pipeline/etl_pipeline.py
```

👉 Now your pipeline runs in cloud


## Automate with Cron

```bash
crontab -e
```

Add:

```cron
0 * * * * python3 /home/ubuntu/the-data-engineer/projects/batch_pipeline/etl_pipeline.py
```

👉 Runs every hour


## Logging in Cloud

```bash
python3 etl_pipeline.py > pipeline.log 2>&1
```

👉 Logs output for debugging
