# Building a Data Lake with AWS S3


## Why Data Lakes Matter

Databases store structured data.

But real-world data is:
- Large
- Unstructured
- Semi-structured

👉 S3 acts as your data lake

Used for:
- Raw data storage
- Processed data
- Analytics datasets


## Data Lake Concept

A data lake stores:

- Raw data (as-is)
- Cleaned data
- Processed data

Structure:

raw/
processed/
analytics/

👉 This is industry standard


## Setup AWS CLI

Install:

```bash
pip install awscli
```

Configure:

```bash
aws configure
```

Enter:
- Access key
- Secret key
- Region


## Create S3 Bucket

```bash
aws s3 mb s3://your-data-engineering-bucket
```


## Upload File

```bash
aws s3 cp sample.json s3://your-data-engineering-bucket/raw/sample.json
```


## Recommended Structure

s3://bucket/

raw/
processed/
analytics/

Example:
raw/api_data.json
processed/cleaned_data.parquet
analytics/report.csv


## Python Upload (Boto3)

```python
import boto3

s3 = boto3.client('s3')

s3.upload_file(
    "sample.json",
    "your-data-engineering-bucket",
    "raw/sample.json"
)

print("Uploaded successfully")
```


## Save Pipeline Output to S3

```python
import json

data = {"message": "hello cloud"}

with open("output.json", "w") as f:
    json.dump(data, f)

s3.upload_file("output.json", "your-data-engineering-bucket", "processed/output.json")
```
