## Working with Data Structures


```python
data = [
    {"id": 1, "name": "Amar", "score": 90},
    {"id": 2, "name": "Ravi", "score": 85},
    {"id": 3, "name": "Neha", "score": 95}
]

# Extract names
names = [d["name"] for d in data]
print(names)

# Filter high scores
top = [d for d in data if d["score"] > 90]
print(top)
```

## Functions (Reusable Logic)


```python
def calculate_average(data):
    total = sum(d["score"] for d in data)
    return total / len(data)

print(calculate_average(data))
```

## Real Data Engineer Pattern


```python
def transform_data(records):
    return [
        {
            "id": r["id"],
            "name": r["name"].upper(),
            "passed": r["score"] > 80
        }
        for r in records
    ]

print(transform_data(data))
```
