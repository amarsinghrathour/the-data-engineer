# File Handling and JSON for Data Engineering


## Why File Handling Matters

In real systems, data does NOT come as Python lists.

It comes as:
- JSON files
- CSV files
- Logs
- API responses

👉 If you can't read/write files, you can't build pipelines.


## Writing to a File


```python
data = "Hello Data Engineering"

with open("sample.txt", "w") as f:
    f.write(data)

print("File written successfully")
```

## Reading from a File


```python
with open("sample.txt", "r") as f:
    content = f.read()

print(content)
```

## Working with JSON (write)


```python
import json

data = {
    "name": "Amar",
    "role": "Data Engineer",
    "skills": ["Python", "SQL"]
}

# Write JSON
with open("data.json", "w") as f:
    json.dump(data, f)

print("JSON saved")
```

## Reading JSON


```python
with open("data.json", "r") as f:
    loaded = json.load(f)

print(loaded)
print(loaded["skills"])
```

## Real-World Pattern


```python
def read_json_file(file_path):
    with open(file_path, "r") as f:
        return json.load(f)


def write_json_file(file_path, data):
    with open(file_path, "w") as f:
        json.dump(data, f, indent=4)


data = read_json_file("data.json")
data["active"] = True

write_json_file("updated.json", data)
```

## Handling Bad Data


```python
try:
    with open("missing.json", "r") as f:
        data = json.load(f)
except FileNotFoundError:
    print("File not found!")
```
