## Standard Pipeline Structure

Every pipeline follows:

1. **Extract** → Get data
2. **Transform** → Clean/process
3. **Load** → Store/output

👉 **ETL** = Extract → Transform → Load


### Logging setup

Run this once per kernel session **before** the ETL steps so `INFO` logs appear.


```python
import logging

logging.basicConfig(level=logging.INFO, format="%(levelname)s - %(message)s")
```

## Step 1: Extract module


```python
import requests
import logging


def extract(url):
    logging.info("Extracting data...")

    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        return response.json()

    except Exception as e:
        logging.error(f"Extraction failed: {e}")
        return []
```

## Step 2: Transform module


```python
def transform(data):
    logging.info("Transforming data...")

    return [
        {
            "id": d.get("id"),
            "title": d.get("title", "").upper(),
            "length": len(d.get("title", "")),
        }
        for d in data
    ]
```

## Step 3: Load module


```python
import json


def load(data, file_name="output.json"):
    logging.info("Loading data...")

    with open(file_name, "w") as f:
        json.dump(data, f, indent=4)

    logging.info(f"Data saved to {file_name}")
```

## Pipeline runner


```python
def run_pipeline(config):
    logging.info("Pipeline started")

    data = extract(config["url"])

    if not data:
        logging.error("No data extracted")
        return

    transformed = transform(data)

    load(transformed, config["output_file"])

    logging.info("Pipeline completed")
```

## Config-driven execution


```python
config = {
    "url": "https://jsonplaceholder.typicode.com/posts",
    "output_file": "final_output.json",
}

run_pipeline(config)
```

## Add timing (production touch)


```python
import time


def run_pipeline_with_timing(config):
    start = time.time()

    run_pipeline(config)

    end = time.time()

    print(f"Pipeline executed in {end - start:.2f} seconds")


run_pipeline_with_timing(config)
```
