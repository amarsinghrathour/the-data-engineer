# Batch vs Streaming Systems


## Why This Matters

Not all data is processed the same way.

Some systems:
- Process data every few hours (batch)
- Process data instantly (streaming)

Choosing the wrong approach can:
- Increase cost
- Slow down systems
- Break real-time features

👉 This is a core decision every Data Engineer makes.


## What is Batch Processing?

Batch processing means:
- Collect data over time
- Process it in chunks

Example:
- Daily sales report
- Nightly ETL jobs

Flow:
Data → Stored → Processed later


```python
import pandas as pd

# Simulating batch data (daily orders)
data = [
    {"order_id": 1, "amount": 100},
    {"order_id": 2, "amount": 200},
    {"order_id": 3, "amount": 150},
]

df = pd.DataFrame(data)

total_sales = df["amount"].sum()

print(f"Total Sales: {total_sales}")

```

## What is Streaming?

Streaming means:
- Data is processed in real-time
- As soon as it arrives

Example:
- Uber ride tracking
- Fraud detection
- Live notifications

Flow:
Data → Process immediately → Output


```python
import time

data_stream = [100, 200, 150, 300]

for value in data_stream:
    print(f"Processing value: {value}")
    time.sleep(1)

```

## Batch vs Streaming

| Feature | Batch | Streaming |
|--------|------|----------|
| Speed | Slow | Real-time |
| Cost | Lower | Higher |
| Complexity | Simple | Complex |
| Use Case | Reports | Live systems |


## Real-World Examples

**Batch:**

**Streaming:**
- Stock market updates
