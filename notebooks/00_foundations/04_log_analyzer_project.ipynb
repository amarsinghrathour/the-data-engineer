# Project: CLI Log Analyzer

**Phase 0 deliverable** — a terminal-first tool that parses logs, aggregates severity, and surfaces failures (the same shape as triaging Spark, Airflow, or service logs).


## Why this matters

Production systems emit **structured or semi-structured** logs. Data engineers:

- **grep** and **count** to estimate blast radius  
- Parse lines into **records** for dashboards or alerts  
- Ship **small CLIs** that others can run during incidents  

This project practices all three at once.


## Repository layout

```text
projects/batch_pipeline/log_analyzer/
├── logs.txt       # sample input
├── analyzer.py    # CLI: argv, printing, report
├── utils.py       # parse_logs(), count_levels()
└── README.md
```

**Convention:** each line looks like `LEVEL message…` (first token is the level).


## Code walkthrough

### `utils.parse_logs`

- Reads the file line by line (memory-friendly for large logs).  
- Splits on the **first space** → `level` + remainder `message`.  

### `utils.count_levels`

- One pass over parsed rows → frequency map by level.

### `analyzer.py`

- Validates `sys.argv`, loads logs, prints summary.  
- **Upgrades in this course version:**  
  1. **Error rate** — `ERROR` count ÷ total lines.  
  2. **`report.txt`** — durable snapshot for tickets or CI artifacts.  
  3. **Optional level filter** — `python analyzer.py logs.txt ERROR` lists only that level’s messages.


## Run (from project folder)

```bash
cd projects/batch_pipeline/log_analyzer
python analyzer.py logs.txt
python analyzer.py logs.txt ERROR
```

Inspect `report.txt` after a run.
