# Linux Basics for Data Engineers


## Why Linux Matters

Most data systems run on Linux servers.

If you can't:
- Navigate files
- Process logs
- Run scripts

You will struggle in real projects.

👉 Data Engineers spend a LOT of time in the terminal.


## Essential Linux Commands

| Command | Description |
|--------|------------|
| `pwd` | Show current directory |
| `ls` | List files |
| `cd` | Change directory |
| `mkdir` | Create folder |
| `rm` | Delete file |
| `cp` | Copy file |
| `mv` | Move file |


## Hands-On (Simulated in Python)

Jupyter is not a full Linux shell, so we simulate filesystem actions with `os`. **Still run the real commands below in your terminal**—that's where the skill sticks.


````python
import os

print("Current Directory:", os.getcwd())

# Create a folder
os.makedirs("test_folder", exist_ok=True)

# List files
print("Files:", os.listdir())
````

## Run These in Terminal (Not Jupyter)

Open **Terminal** (macOS/Linux) or **WSL** (Windows). Run:

```bash
pwd
ls
mkdir data_engineering
cd data_engineering
touch sample.txt
echo "hello world" > sample.txt
cat sample.txt
```


## File Operations

**Still in your terminal** (inside `data_engineering` if you followed above):

```bash
cp sample.txt copy.txt
mv copy.txt moved.txt
rm moved.txt
```


## Log Processing (Important Skill)

```bash
cat sample.txt
wc -l sample.txt
grep "hello" sample.txt
```


## Python + Linux Combo

Call shell commands from Python when you need scripting glue (CI, small tools, Airflow hooks):


````python
import subprocess

result = subprocess.run(["ls"], capture_output=True, text=True)
print(result.stdout)
````
