# Apache Spark Basics for Data Engineering


## Why Spark Matters

When data becomes large:
- Python + Pandas breaks
- Memory becomes a bottleneck

Spark solves this by:
👉 **Distributing** work across a cluster

Used by:
- Netflix
- Amazon
- Uber


## What is Apache Spark?

Spark is a **distributed** data processing engine.

Instead of:
👉 One machine holding all data in RAM

You get:
👉 Many executors processing **partitions** in parallel

## Create Spark session


```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder.appName("SparkIntro")
    .master("local[*]")
    .getOrCreate()
)

spark
```

## Create DataFrame


```python
data = [
    (1, "Amar", 100),
    (2, "Ravi", 200),
    (3, "Neha", 300),
]

columns = ["id", "name", "amount"]

df = spark.createDataFrame(data, columns)

df.show()
```

## Basic operations


```python
df.select("name").show()

df.filter(df["amount"] > 150).show()

df.groupBy("name").sum("amount").show()
```

## Key concepts

**Lazy evaluation:** transformations build a **plan** — nothing executes until an **action**.

**Transformations** (lazy): `filter`, `select`, `join`, …

**Actions** (eager): `show()`, `collect()`, `count()`, `write`, …

👉 Work runs when you trigger an action.
