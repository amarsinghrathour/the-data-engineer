# Spark SQL for Data Engineering


## Why Spark SQL Matters

You already know SQL.

Now imagine:
👉 Running **the same patterns** on **millions** (or billions) of rows

**Spark SQL** compiles SQL into the same engine as the DataFrame API.

Used for:
- Analytics
- Batch pipelines
- Reporting / BI extracts


## Prerequisites

- PySpark + JDK ([`README.md`](README.md)).


## Spark session


```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder.appName("SparkSQLIntro")
    .master("local[*]")
    .getOrCreate()
)
```

## Create DataFrame


```python
data = [
    (1, "Amar", 100),
    (1, "Amar", 200),
    (2, "Neha", 300),
    (2, "Neha", 150),
    (3, "Ravi", 400),
]

columns = ["user_id", "name", "amount"]

df = spark.createDataFrame(data, columns)
df.show()
```

## Register temporary view

`createOrReplaceTempView` exposes the DataFrame as a SQL table for this Spark session.


```python
df.createOrReplaceTempView("transactions")
```

## Basic SQL query


```python
result = spark.sql(
    """
SELECT name, SUM(amount) as total
FROM transactions
GROUP BY name
"""
)

result.show()
```

## Filtering + aggregation


```python
spark.sql(
    """
SELECT name, COUNT(*) as total_transactions
FROM transactions
WHERE amount > 100
GROUP BY name
"""
).show()
```

## Join example


```python
users = [
    (1, "Amar"),
    (2, "Neha"),
    (3, "Ravi"),
]

users_df = spark.createDataFrame(users, ["user_id", "name"])
users_df.createOrReplaceTempView("users")

spark.sql(
    """
SELECT u.name, SUM(t.amount) as total_spent
FROM transactions t
JOIN users u ON t.user_id = u.user_id
GROUP BY u.name
"""
).show()
```

## Window function (Spark SQL)


```python
spark.sql(
    """
SELECT *,
RANK() OVER (PARTITION BY user_id ORDER BY amount DESC) as rank
FROM transactions
"""
).show()
```

## Real pipeline query


```python
spark.sql(
    """
SELECT name, SUM(amount) as total_spent
FROM transactions
GROUP BY name
ORDER BY total_spent DESC
"""
).show()
```
